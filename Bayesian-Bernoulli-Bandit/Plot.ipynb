```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
T = 15
plcy_name = ["fp", "ucb", "ts"]
dfs = [pd.read_csv(name + "-" + str(T), index_col=0) for name in plcy_name]
colors = ["r", "g", "b"]
policy_names = ["Fluid-Priority", "UCB", "TS"]

ax = plt.subplot(111)
ax.set_xscale("log", basex=2)
ax.set_yscale("log", basey=2)
ax.set_ylim((0.03, 1000))
ax.set_xlim((100, 40000))
for i in range(3):
    plt.errorbar(dfs[i].loc["N"], dfs[i].loc["opt-gap"], dfs[i].loc["std"] * 2, marker="o",
                 color=colors[i], ls=":", label=policy_names[i])
plt.legend()
plt.xlabel("Number of arms")
plt.ylabel("Optimality gap's upper bound")
plt.title(f"Bayesian Bernoulli Bandit (T = {T})")
plt.grid()
plt.savefig("1a.png")
```
