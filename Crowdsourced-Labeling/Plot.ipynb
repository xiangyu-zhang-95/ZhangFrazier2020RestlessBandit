```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
fnames = ["fp.csv", "kg.csv", "okg.csv"]
plcies = ["Fluid-Priority", "Knowledge-Gradient", "Optimistic Knowledge-Gradient"]
dfs = [pd.read_csv(fname, index_col=0) for fname in fnames]
clrs = ["r", "b", "g"]
```

```python
for i in range(3):
    plt.errorbar(dfs[i].loc["N"][1:], dfs[i].loc["opt-gap"][1:], marker="o", ls=":",
                 yerr=2 * dfs[i].loc["std"][1:], label=plcies[i], color=clrs[i])
plt.xlabel("Number of arms")
plt.ylabel("Optimality Gap's Upper Bound")
plt.legend()
plt.grid()
plt.savefig("2a.png")
```

```python
for i in range(3):
    plt.errorbar(dfs[i].loc["N"][1:], dfs[i].loc["opt-gap"][1:], marker="o", ls=":",
                 yerr=2 * dfs[i].loc["std"][1:], label=plcies[i], color=clrs[i])
plt.xlabel("Number of arms")
plt.ylabel("Optimality gap's upper bound")
plt.ylim(0, 2)
plt.legend()
plt.grid()
plt.savefig("2b.png")
```
